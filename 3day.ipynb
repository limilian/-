```python
import torch
# 定义sigmoid函数
def sigmoid(x):
    return 1 / (1 + torch.exp(-x))

# 定义sigmoid的导数函数
def sigmoid_derivative(x):
    y = sigmoid(x)
    return y * (1 - x)

# 在x=0处计算sigmoid值及其导数
x = torch.tensor(0.0, requires_grad=True)  # 创建一个可求导的张量
sigmoid_val = sigmoid(x)
sigmoid_deriv = sigmoid_derivative(x)

# 使用autograd自动计算梯度
sigmoid_val.backward()  # 这会计算sigmoid_val对z的梯度，并存储在z.grad中

# 打印结果
print(f"Sigmoid at x=0: {sigmoid_val.item()}")
print(f"Derivative of Sigmoid at x=0 (manual calculation): {sigmoid_deriv.item()}")
print(f"Derivative of Sigmoid at x=0 (autograd): {x.grad.item()}")  # 这应该与sigmoid_deriv.item()相同
```

```python
import torch

# 定义输入变量
x1 = torch.tensor(0.0, requires_grad=True)  # x1 = 0
x2 = torch.tensor(1.0, requires_grad=True)  # x2 = 1

# 定义模型
def model(x1, x2):
    term1 = torch.sigmoid(x1 + 2*x2 + 1)
    term2 = torch.sigmoid(x1 + (2*x2) + 1)
    y = torch.sigmoid(3 * term1 + term2 + 1)
    return y

# 计算模型
y = model(x1, x2)

# 计算梯度
y.backward()

# 打印梯度
print(f" x1 梯度 is: {x1.grad.item()}")
print(f" x2 梯度 is: {x2.grad.item()}")
```
